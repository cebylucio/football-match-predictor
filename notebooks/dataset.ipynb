```python

# used for working with data sets 
import pandas as pd
# used for file system operations 
import os 

```

Loading of data per year

```python

urls = [
    "https://www.football-data.co.uk/mmz4281/2526/E0.csv",
    "https://www.football-data.co.uk/mmz4281/2425/E0.csv",
    "https://www.football-data.co.uk/mmz4281/2324/E0.csv",
    "https://www.football-data.co.uk/mmz4281/2223/E0.csv",
    "https://www.football-data.co.uk/mmz4281/2122/E0.csv",
]

season_data = []
# load all data year by year
for i in range(len(urls)):
    print(f"downlaoding season {i + 1} of {len(urls)}")
    hold  = pd.read_csv(urls[i])
    season_data.append(hold)
    
df = pd.concat(season_data, ignore_index=True)

df.to_csv('raw_data.csv', index=False)


print("done")
print("Total matches", len(df))
print("Number of columns", df.shape[1])
    
```

```python

# FTHG = full time home goal , FTAG = full time away goal , FTR = full time result , hs = home shots , as = away shots , hst = Home Shots on Target , ast  = Away Shots on Target
columns_needed = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST']


df = df[columns_needed].copy()

# get date for sorting 
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# sort the dates 
df = df.sort_values(by='Date')

```

```python

# sort the dates
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.sort_values(by='Date').reset_index(drop=True)

# data in order 
df.to_csv('sorted_data.csv', index=False)


print(f"Final match count: {len(df)}")
print(f"Date range: {df['Date'].min().date()} to {df['Date'].max().date()}")
```
